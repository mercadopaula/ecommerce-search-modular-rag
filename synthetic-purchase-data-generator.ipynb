```python
import random
import datetime

# Function to generate a random date in 2023 or 2024
def random_date():
    year = random.choice([2023, 2024])
    month = random.randint(1, 12)
    day = random.randint(1, 28)  # Keeping it safe for all months
    return datetime.date(year, month, day).isoformat()

# Function to generate a random product
def generate_product():
    return {
        "product_id": random.randint(10, 700),
        "price": round(random.uniform(50, 1000), 2),  # Price in Euros
        "quantity": 1  # Always 1 as per requirement
    }

# Function to generate a purchase order
def generate_purchase_order():
    return {
        "purchase_date": random_date(),
        "products": [generate_product() for _ in range(random.randint(1, 5))]
    }

# Function to generate a customer with purchase history
def generate_customer(customer_id):
    return {
        "customer_id": customer_id,
        "purchase_orders": [generate_purchase_order() for _ in range(random.randint(0, 8))]
    }

# Generate a list of 10 customers
customers = [generate_customer(customer_id=i) for i in range(1, 11)]

# Print the generated data
import json
print(json.dumps(customers, indent=4))
```
